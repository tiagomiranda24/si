```python
import os
import numpy as np
from si.data.dataset import Dataset
from si.io.csv_file import read_csv
from si.decomposition.pca import PCA
```

Test the PCA class in a jupyter notebook using
the iris.csv dataset (classificação).

```python
path = r"C:\Users\tiago\OneDrive\Documentos\GitHub\si\datasets"
csv_file = os.path.join(path, 'iris', 'iris.csv')
dataset = read_csv(filename=csv_file, features=True, label=True)
```

```python
# Create an instance of PCA with 2 components
pca = PCA(n_components=2)

# Fit the PCA model to the data and transform it
pca.fit(dataset)
```

```python
# Transform the data into the new principal component space
X_reduced = pca.transform(dataset)

print("Transformed data (reduced X):\n", X_reduced)
```

```python
# Display the principal components
print("Principal components:\n", pca.components)

# Display the explained variance for each component
print("Explained variance for each component:\n", pca.explained_variance)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot of the reduced data, colored by labels
plt.figure(figsize=(8, 6))
for label in np.unique(dataset.y):
    plt.scatter(X_reduced[dataset.y == label, 0], 
                X_reduced[dataset.y == label, 1], 
                label=f"Class {label}")

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA - Iris Dataset')  
plt.legend()  
plt.show()  
```
